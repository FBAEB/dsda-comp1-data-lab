# Database to Visualisation

This is a super simple example of a pipeline pulling data from a relational database and plotting it on a pie chart.  By doing this you are pushing some processing to the database engine, you then incur some network overhead as the data moves across the network to the machine running your python code.  Ideally you want to be minimising the network traffic.

```python
import pandas as pd
import mysql.connector
import matplotlib.pyplot as plt
```

```python
cnx = mysql.connector.connect(user="root",
                              password="Nott1ngham",
                              host="localhost",
                              port=3306,
                              database="tyres",
                              use_pure=True)
```

```python
sql = 'SELECT Size, Count(Model) as Frequency from tyres group by tyres.Size order by Frequency desc'
```

```python
df = pd.read_sql(sql, cnx)
```

```python
fig1, ax1 = plt.subplots()
chart = plt.pie(df['Frequency'].tolist(), labels=df['Size'].tolist())
```
